```python
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=50)
```

```python
roll_belt=dfTraining['roll_belt'].values.reshape(-1,1)
pitch_belt=dfTraining['pitch_belt'].values.reshape(-1,1)
```

```python
belt=np.column_stack((roll_belt,pitch_belt))
print(belt)
```

```python
roll_arm=dfTraining['roll_arm'].values.reshape(-1,1)
pitch_arm=dfTraining['pitch_arm'].values.reshape(-1,1)

arm=np.column_stack((roll_arm,pitch_arm))
print(arm)
```

```python
roll_dumbbell=dfTraining['roll_dumbbell'].values.reshape(-1,1)
pitch_dumbbell=dfTraining['pitch_dumbbell'].values.reshape(-1,1)

dumbbell=np.column_stack((roll_dumbbell,pitch_dumbbell))
print(dumbbell)
```

```python
roll_forearm=dfTraining['roll_forearm'].values.reshape(-1,1)
pitch_forearm=dfTraining['pitch_forearm'].values.reshape(-1,1)

forearm=np.column_stack((roll_forearm,pitch_forearm))
print(forearm)
```

```python
classification=dfTraining['classe'].values.reshape(-1,1)
classe=np.column_stack((classification))
print(classe)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
model_2=MLPClassifier(hidden_layer_sizes=(100,100,100), max_iter=500, alpha=0.0001, solver='sgd', verbose=10, random_state=50, tol=0.000000001)
```

```python
model_2.fit(X_train,Y_train)
Y_pred=model_2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, Y_pred)
```
